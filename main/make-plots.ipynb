```python
import modelframe as mf
import models as ms
import numpy as np
import matplotlib.pyplot as plt
import plotutilities as pu
import datautilities as du
```

```python
datadir = "../model/model-binary-weighted-001"

frame = mf.GlitchModel(ms.GlitchClassifier(), None, None)
frame.read_result_data(datadir)

fig,ax = pu.plot_performance(frame)
fig.savefig(f"{datadir}/figures/performance-overview-1.pdf", bbox_inches = "tight")
fig.savefig(f"{datadir}/figures/performance-overview-1.png", bbox_inches = "tight")

fig, ax = pu.plot_recall_matrix(frame)
fig.savefig(f"{datadir}/figures/recall-matrix-1.pdf", bbox_inches = "tight")
fig.savefig(f"{datadir}/figures/recall-matrix-1.png", bbox_inches = "tight")

fig, ax = pu.plot_precision_matrix(frame)
fig.savefig(f"{datadir}/figures/precision-matrix-1.pdf", bbox_inches = "tight")
fig.savefig(f"{datadir}/figures/precision-matrix-1.png", bbox_inches = "tight")

fig, ax = pu.plot_accuracy_matrix(frame)
fig.savefig(f"{datadir}/figures/accuracy-matrix-1.pdf", bbox_inches = "tight")
fig.savefig(f"{datadir}/figures/accuracy-matrix-1.png", bbox_inches = "tight")


```
